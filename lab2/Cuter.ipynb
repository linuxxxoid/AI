### Импортируем библиотеки

```python
import cv2
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['figure.figsize'] = (10,10)
```

### Импортируем изображение

```python
im = cv2.imread("data/source/max.jpg")

plt.imshow(im)

print("Размер изображения:", str(im.shape[0]) + 'x' + str(im.shape[1]))
```

На изображении 15 колонок и 21 строка. По 7 строк на каждый из классов.

Будем брать изображения по 32 пикселя с отступом в 2 пикселя в надежде на то что мы минимизируем присутствие разлиновки на изображениях.

```python
step = 34
line_num = 15
col_num = 21

step_by_class = 7

classes = ['ugly', 'bad', 'good']
```

### Изменим размер изображения

```python
need_height = col_num*step
print("Необходимый размер в высоту:", need_height)
```

```python
need_weight = line_num*step
print("Необходимый размер в ширину:", need_weight)
```

```python
resized = cv2.resize(im, (need_weight, need_height), interpolation = cv2.INTER_AREA)
```

```python
plt.imshow(resized)

print("Размер изображения:", str(resized.shape[0]) + 'x' + str(resized.shape[1]))
```

### Вырежем изображения

Примерно так будет выглядеть изображение при нарезке. Не получилось идеально избавиться от разлиновки на изображениях, однако глупо надеятся, что данные, с которыми в теории будет работать нейросеть будут идеально чистыми.

```python
plt.imshow(resized[1:step-1, 1:step-1])
```

Нарезка с отступом в 2 пикселя и сохранением в отдельные папки для каждого класса:

```python
for n in np.arange(len(classes)):
    for i in np.arange(n * step * step_by_class, (n + 1) * step * step_by_class, step):
        for j in np.arange(0, need_weight, step):
            cut = resized[i + 1 : i + step - 1, j + 1 : j + step - 1]
            cv2.imwrite("data/max/" + classes[n] + "/img_" + str(i//step) + "x" + str(j//step) + ".jpg", cut)
```

#### Первый датасет готов!
